```python
######problem1--knapsack multiple optima######
#########create weights#####
import mlrose_hiive as mlrose
import numpy as np
import time
import random
import matplotlib.pyplot as plt
length =[5,8,11,14]
weights_list =[]
for length in length:
    index_length = 5
    index_list = np.array(random.sample(range(length),index_length))
    weights = np.zeros(length)
    weights[index_list] = np.array(random.choices([1,2],[2,1],k=index_length))
    weights_list.append(weights)

print(weights_list)

###create the fitness function########
def cust_fn(state, w, c): 
    return -(np.dot(state,w)-c)**2

###Compare the four algorithm on the problem####
RHL_time,RHL_Accuracy,RHL_iteration =[],[],[]
SA_time, SA_Accuracy,SA_iteration = [],[],[]
GA_time,GA_Accuracy,GA_iteration = [],[],[]
MIMIC_time,MIMIC_Accuracy,MIMIC_iteraton = [],[],[]
# weights_list = [np.array([1,2,4,9,19,40,77,117])]
for weights in weights_list:
    #define fitness function####
    c = 2
    print(c)
    kwargs = {'w':weights,'c': c}
    fitness = mlrose.CustomFitness(cust_fn, **kwargs)
    #create the problem 
    problem = mlrose.DiscreteOpt(length = weights.shape[0], fitness_fn = fitness, max_val = 2)
    #use randomized_hill_climbing####
    total_time = 0
    average_accuracy = 0
    iteration = 0
    for i in range(10):   
        problem.reset()
        start_time = time.time()
        best_state,best_fitness,fitness_curve=mlrose.random_hill_climb(problem, max_attempts=length, max_iters=np.inf, restarts=0,init_state=None, curve=True, random_state=None)
        total_time += (time.time()-start_time)*1000
        iteration += len(fitness_curve)
        if best_fitness==0: average_accuracy +=1
    RHL_time.append(total_time/10) 
    RHL_Accuracy.append(average_accuracy/10)
    RHL_iteration.append(iteration/10)
    #use SA####
    total_time = 0
    average_accuracy = 0
    iteration = 0
    for i in range(10):   
        problem.reset()
#         schedule=mlrose.GeomDecay(init_temp=10, decay=0.95, min_temp=0.001)
        schedule = mlrose.ArithDecay(init_temp=1.0, decay=0.95, min_temp=0.1)
        start_time = time.time()
        best_state,best_fitness,fitness_curve=mlrose.simulated_annealing(problem, schedule=schedule, max_attempts=length, max_iters=np.inf, init_state=None, curve=True,random_state=None)
        total_time += (time.time()-start_time)*1000
        iteration += len(fitness_curve)
        if best_fitness==0: 
            average_accuracy +=1
    SA_time.append(total_time/10) 
    SA_Accuracy.append(average_accuracy/10)
    SA_iteration.append(iteration/10)
    ###use GA###
    total_time = 0
    average_accuracy = 0
    iteration = 0
    for i in range(10):   
        problem.reset()
        start_time = time.time()
        best_state,best_fitness,fitness_curve=mlrose.genetic_alg(problem, pop_size=200, mutation_prob=0.1, max_attempts=10,max_iters=np.inf, curve=True, random_state=None)
        total_time += (time.time()-start_time)*1000
        iteration += len(fitness_curve)
        if best_fitness==0: 
#             print("found")
            average_accuracy +=1
    GA_time.append(total_time/10) 
    GA_Accuracy.append(average_accuracy/10)
    GA_iteration.append(iteration/10)
    ###use MIMIC###
    total_time = 0
    average_accuracy = 0
    iteration = 0
    for i in range(10):   
        problem.reset()
        start_time = time.time()
        best_state,best_fitness,fitness_curve=mlrose.mimic(problem, pop_size=200, keep_pct=0.2, max_attempts=10, max_iters=np.inf, curve=True, random_state=None)
        total_time += (time.time()-start_time)*1000
        iteration += len(fitness_curve)
        if best_fitness==0: 
            average_accuracy +=1
    MIMIC_time.append(total_time/10) 
    MIMIC_Accuracy.append(average_accuracy/10)
    MIMIC_iteration.append(iteration/10)
        
W=['W_1','W_2','W_3','W_4']
fig_0, axs = plt.subplots(1, 1, figsize=(5,5), sharey=False)
axs.plot(W,RHL_time,"g^",linestyle = "--",label='RHL')
axs.plot(W,SA_time,"ro",linestyle = "--",label='Simulated Annelling')
axs.plot(W,GA_time,"b^",linestyle = "--",label='Genetic Algorithm')
axs.plot(W,MIMIC_time,"yo",linestyle = "--",label='MIMIC')
axs.set_xlabel("Knapsack Weights")
axs.set_ylabel("Time")
axs.legend()
fig_0.suptitle("problem1:Time by four algorithms (million_seconds)")
fig_0.savefig("problem1:Time by four algorithms (million_seconds)")

fig_1,axs = plt.subplots(1, 1, figsize=(5, 5), sharey=False)
axs.plot(W,RHL_Accuracy,"g^",linestyle = "--",label='RHL')
axs.plot(W,SA_Accuracy,"ro",linestyle = "--",label='Simulated Annelling')
axs.plot(W,GA_Accuracy,"b^",linestyle = "--",label='Genetic Algorithm')
axs.plot(W,MIMIC_Accuracy,"yo",linestyle = "--",label='MIMIC')
axs.set_xlabel("Knapsack Weights")
axs.set_ylabel("Accuracy")
plt.ylim(-1,1.5)
axs.legend()
fig_1.suptitle("problem1:Accuracy by four algorithms")
fig_1.savefig("problem1:Accuracy by four algorithms.png")

fig_2,axs = plt.subplots(1, 1, figsize=(5, 5), sharey=False)
axs.plot(W,RHL_iteration,"g^",linestyle = "--",label='RHL')
axs.plot(W,SA_iteration,"ro",linestyle = "--",label='Simulated Annelling')
axs.plot(W,GA_iteration,"b^",linestyle = "--",label='Genetic Algorithm')
axs.plot(W,MIMIC_iteration,"yo",linestyle = "--",label='MIMIC')
axs.set_xlabel("Knapsack Weights")
axs.set_ylabel("Iteration Numbers")
axs.legend()
fig_2.suptitle("problem1:iteration numbers y four algorithms")
fig_2.savefig("problem1:iteration numbers by four algorithms.png")
```
